```python
import pandas as pd
import numpy as np
#make df
df = pd.read_excel(r'D:\Function App Files\Open Opportunities Funnel-Jan22.xlsx',header = 1)
display(df)
```

```python
#transpose on df and store to df
df = df.transpose()
display(df)
```

```python
display(df.iloc[1])
```

```python
#regex replace Space To UnderScore
df.iloc[1] = df.iloc[1].replace(' ', '_', regex=True)
```

```python
#drop null
df = df.dropna(how='all')
display(df)
```

```python
#make Second Row AS Header 
df.columns = df.iloc[0]
df = df[1:]
df.head()
display(df)
```

```python
#Drop Null Columns
df = df.loc[:, df.columns.notnull()]
display(df)
```
